# The Pythagorean Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Math,display
```

* The Pythagorean Theorem
* How to create triangles in Python
* An application of the Pythagorean Theorem to complex numbers

$c^2 = a^2 + b^2$

$ c = \sqrt{a^2 + b^2} $

### Pythagorean Triangle Generator

```python
a = 3
b = 4
c = np.sqrt(a**2+b**2)

# triangle
plt.plot([0,a],[0,0],'k')
plt.plot([0,0],[0,b],'k')
plt.plot([0,a],[b,0],'k')

plt.plot([.3,.3],[0,.3],'k')
plt.plot([0,.3],[.3,.3],'k')

plt.text(a/3,-.35,'a='+str(a),fontsize=15)
plt.text(-.8,b/2.75,'b='+str(b),fontsize=15)
plt.text(a/2+.2,b/2,'c='+str(int(c)),fontsize=15)

plt.axis('square')
axlim = max(a,b) + .5
plt.axis([-.5,axlim,-.5,axlim])
# plt.grid()
plt.axis('off')
plt.show()
```

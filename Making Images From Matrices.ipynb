# Making Images From Matrices

```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
from IPython.display import Math,display
```

```python
display(Math('\\text{equation = } A_{i,j} = 3 \\times i - 4 \\times j'))
```

```python
A = [ [1,2],[1,4] ]

plt.imshow(A)

plt.xticks([0,1])
plt.yticks([0,1])

plt.show()
```

```python
A = np.zeros((10,14))
print(A)
print('\n')
print( np.shape(A) )
```

```python
for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i + 4*j
```

```python
plt.imshow(A)

plt.show()
```

```python
for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i * 4*j
```

```python
plt.imshow(A)

plt.show()
```

```python
for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i - 4*j

plt.imshow(A)
plt.plot([0,2],[8,2],'w-',linewidth=4)
        
for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center',c='k')

plt.set_cmap('Reds')
plt.savefig('RedPlotFigure.pdf')
plt.show()

```

### Exercise

```python
display(Math('\\text{equation = } C_{i,j} = -1^{i+j}'))
```

```python
# initialize a new matrix with soft coding
N = 10
C = np.zeros((N,N))
print(C)
```

```python
# define the equation
for i in range(0,N):
    for j in range(0,N):
        C[i,j] = (-1)**(i+j)
```

```python
print(C)
```

```python
plt.tick_params(labelleft=False,labelbottom=False)
plt.tick_params(left=False,bottom=False)
plt.imshow(C)
plt.show()
```

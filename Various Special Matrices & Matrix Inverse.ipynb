# Matrix Inverse

* Why a "matrix inverse" is necessary
* The conditions for an inverse to exist 
* How to use and interpret the pseudoinverse

## $ 3x = 1 $

### Scalar Inverse

## $ 3^{-1} 3x = 3^{-1}1 $

## $ 1x = 3^{-1} $

## Matrix Inverse Example

## $ Ax = b $

## $ A^{-1} Ax = A^{-1}b $

## $ Ix = A^{-1}b $

## $ x = A^{-1}b $

**Matrix is invertible if it is:**
* Sqaure 
* Full rank (Every column provides unique information in the matrix)

```python
m = 5
A = np.random.randint(-5,6,(m,m))
Ainv = np.linalg.inv(A)

print(A)
print('\n')
print(Ainv)
```

```python
fig,ax = plt.subplots(1,3)

ax[0].imshow(A)
ax[0].set_title('A \n (Original Matrix)')

ax[1].imshow(Ainv)
ax[1].set_title('A$^{-1}$ \n (Matrix Inverse)')

ax[2].imshow(A@Ainv)
ax[2].set_title('A $ \\times $ A$^{-1}$ \n (Identity Matrix)')

for i in ax:
    i.set_xticks([])
    i.set_yticks([])

plt.show()
```

### Matrix Pseudoinverse

**Compute the pseudoinverse and make images of A,A\*, and AA\* for:**
1. A 5x5 matrix with random integers between -5 and +5, where the first column equals the second column
2. A 5x7 matrix of Gaussian random numbers

```python
m = 5
A = np.random.randint(-m,m+1,(m,m))
A[:,0] = A[:,1]
Apinv = np.linalg.pinv(A)

fig,ax = plt.subplots(1,3)
ax[0].imshow(A)
ax[0].set_title('A \n (Matrix)')

ax[1].imshow(Apinv)
ax[1].set_title('A* \n (Matrix Pseudoinv)')

ax[2].imshow(A@Apinv)
ax[2].set_title('A$\\times$A* \n (Multiplied)')

for i in ax:
    # i.set_xticks([])
    # i.set_yticks([])
    i.axis('off')

plt.show()
```

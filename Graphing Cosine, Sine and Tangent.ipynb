# Graphing Cosine, Sine and Tangent

```python
import matplotlib.pyplot as plt 
from IPython.display import Math,display
import sympy as sym
import numpy as np
```

### Importance of Resolution

```python
x = np.linspace(0,6*np.pi,400) # resolution of x in steps

plt.plot(x,np.sin(x),'r',label='sin($\\theta$)')
plt.plot(x,np.cos(x),'b',label='cos($\\theta$)')

plt.title('sine & cosine of $\\theta$')
plt.legend()
plt.xlabel('rad.')
plt.ylabel('f(x) value')
plt.show()
```

```python
# tangent is defined as the ratio between sine and cosine

theta = np.linspace(0,4*np.pi,300)

plt.plot(theta,np.tan(theta),'k')
plt.ylim([-5,5])
plt.title('tangent of $\\theta$')
plt.show()
```

```python
ang = np.random.rand()*2*np.pi

tan = np.tan(ang)
sc = np.sin(ang) / np.cos(ang)

print(ang,tan - sc)
```

```python
thetas = np.linspace(0,2*np.pi,13)
print(thetas)

print(np.cos(thetas)**2 + np.sin(thetas)**2)
```

```python
theta = np.linspace(0,6*np.pi,200)
y1 = np.sin(theta + np.cos(theta))
y2 = np.cos(theta + np.sin(theta))

plt.plot(y1,'r',label='y1')
plt.plot(y2,'b',label='y2')

plt.legend()
plt.show()
```

```python
theta = np.linspace(-2*np.pi,2*np.pi,200)
y1 = np.sin(theta + np.cos(theta))
y2 = np.cos(theta + np.sin(theta))

plt.plot(theta,y1,'r',label='y1 = sin($\\theta$ + cos($\\theta$))')
plt.plot(theta,y2,'b',label='y2 = cos($\\theta$ + sin($\\theta$))')

plt.legend()
plt.show()
```

### Exercise II

```python
theta_sym = sym.pi*7/6
theta = float(theta_sym) 
x = np.linspace(0,np.pi*2,200)

# the unit circle
plt.plot(np.cos(x),np.sin(x),color='gray')
plt.plot([-1.1,1.1],[0,0],'--k',alpha=.3)
plt.plot([0,0],[-1.1,1.1],'--k',alpha=.3)

# separate sine/cosine components
plt.plot([0,0],[0,np.sin(theta)],'r',label='sine segment')
plt.plot([0,np.cos(theta)],[np.sin(theta),np.sin(theta)],'r:')
plt.plot([0,np.cos(theta)],[0,0],'b',label='cosine segment')
plt.plot([np.cos(theta),np.cos(theta)],[0,np.sin(theta)],'b:')

# arc 
plt.plot(np.cos(np.linspace(0,theta,200)),np.sin(np.linspace(0,theta,200)),'k',linewidth=3)
plt.plot([0,np.cos(theta)],[0,np.sin(theta)],'k:')
plt.plot(np.cos(theta),np.sin(theta),'ko')

plt.legend()
plt.title('$\\theta = %s$' %sym.latex(theta_sym))
plt.axis('square')
plt.show()
```

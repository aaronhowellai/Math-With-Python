# Piecewise Functions

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt 
from IPython.display import Math,display
```

# $ f(x) $ = \begin{cases} 0 & \text{if } x \leq 0 \\ -2x & \text{if } x > 0 
\end{cases}

```python
# Define the function
def f(x):
    return np.where(x <= 0, 0, -2 * x)

# Generate x values
x = np.linspace(-5, 5, 400)
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x)', color='blue')
plt.title('Piecewise Function Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-10, 5)
plt.show()
```

```python
from sympy.abc import x

p1 = 0 
p2 = -2*x
p3 = x**3/10

fx = sym.Piecewise(  (p1,x<0),(p2,(x>=0) & (x<3)),(p3,x>=3)  )

fx
```

```python
fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,5,1234)

plt.plot(xx,fxx(xx))
plt.show()
```

# Computing The Limits of a Function

# $ \lim_{x \to a} f(x) $


The limit of f(x), as x goes towards a

# $ \lim_{x \to 4} \frac{x^2}{2} = \frac{4^2}{2} = 8 $


# $ f(x) = \frac{x^2}{x-2} $


# $ f(x) = \frac{x^2}{x-2} $


# $ \lim_{x \to 2} f(x) = \space ?? $


# $ \lim_{x \to 2^+} f(x) = +\infty $


# $ \lim_{x \to 2^-} f(x) = -\infty $


```python
from sympy.abc import x

fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx,x,lim_pnt)

lim
```

```python
display(Math('\\lim_{x \\to %g} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

```python
fxx = sym.lambdify(x,fx)
xx = np.linspace(-5,5,200)

plt.plot(xx,fxx(xx))
plt.plot(lim_pnt,lim,'ro')
plt.show()
```

```python
fx = (x**2) / (x-2)
fxx = sym.lambdify(x,fx)
xx = np.linspace(1,3,102)

lim_pnt = 2
lim = sym.limit(fx,x,lim_pnt,dir='-')

plt.plot(xx,fxx(xx),'r')
display(Math('\\lim_{x \\to %g^-} %s = %g' %(lim_pnt,sym.latex(fx),lim)))

plt.title('$\lim_{x \\to %g^{-}} %s = %g$' %(lim_pnt,sym.latex(fx),lim))
plt.show()
```

# $ f(x) = e^{-x}\sqrt{x+1} $

# $ g(x) = cos(x+sin(x)) $

    My Implementation
    (Correct)

```python
from sympy.abc import x

fx = sym.sqrt(x+1) * sym.exp(-x)
gx = sym.cos(x+sym.sin(x))
xx = np.linspace(0,10,100)

lim_pnt = 5
lim_fx = sym.limit(fx,x,lim_pnt)
lim_gx = sym.limit(gx,x,lim_pnt)

# lambdify 
fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x,gx)

print(lim_fx / lim_gx == sym.limit(fx/gx,x,lim_pnt))

plt.plot(xx,fxx(xx))
plt.plot(xx,gxx(xx),'r')
plt.show()
```

Example Implementation

```python
fx = sym.sqrt(x+1) * sym.exp(-x)
gx = sym.cos(x+sym.sin(x))

# lambdify 
fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x,gx)

xx = np.linspace(0,10,100)

plt.plot(xx,fxx(xx),label='f(x)')
plt.plot(xx,gxx(xx),'r',label='g(x)')
plt.legend()
plt.show()
```

```python
lim_pnt = 5
lim_fx = sym.limit(fx,x,lim_pnt)
lim_gx = sym.limit(gx,x,lim_pnt)

hx = fx/gx
lim_fgx = sym.limit(hx,x,lim_pnt)

display(Math('\\frac{\\lim_{x \\to %g} f(x)}{\\lim_{x \\to %g} g(x)} = \\frac{%g}{%g} = %g' #\
             %(lim_pnt,lim_pnt,lim_fx,lim_gx,lim_fx/lim_gx)))

display(Math('\\lim_{x\\to%g} \\frac{f(x)}{g(x)} = %g' %(lim_pnt,lim_fgx)))
```

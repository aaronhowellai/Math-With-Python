```python
import matplotlib.pyplot as plt
import numpy as np
```

#### My Implementation

```python
t = np.linspace(0,50,500)
a = t
x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m')
plt.axis('square')
plt.show()
```

```python
t = np.linspace(-6*np.pi,6*np.pi,1000)
a = t
x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m')
plt.axis('square')
plt.show()
```

### Exercise: My Implementation
(correct)

### $ a = t^n $

### $ n \in $ {0,1,...,8} 

### $ x = a \space (cos(t)^3) $

### $ y = a \space (sin(t)^3) $

```python
t = np.linspace(-6*np.pi,6*np.pi,1000)

for n in range(0,9):
    a = t**n 
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    plt.plot(x,y,'m')
    plt.axis('square')
    plt.axis('off')
    plt.title('a = $t^{%g}$' %(n))
    
    plt.show()
```

### Example Implementation

```python
fig,ax = plt.subplots(3,3,figsize=(10,8))
row,col = np.indices((3,3))

for n in range(9):
    a = t**n
    x = a * np.cos(t)**3
    y = a * np.sin(t)**3
    
    r = row.ravel()[n]
    c = col.ravel()[n]
    
    ax[r,c].plot(x,y,'k')
    ax[r, c].set_title('a = $t^{%g}$' % n)
    ax[r,c].axis('square')
    ax[r,c].axis('off')
    
    
plt.show()
```

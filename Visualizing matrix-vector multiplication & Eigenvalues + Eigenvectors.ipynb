# Visualizing matrix-vector multiplication

* A geometric interpretation of matrix-vector multiplication
* Build intuition for eigenvectors

```python
M = np.array([ [.5,1],
               [1,.5] ])

v = np.array([1,2])

Mv = M@v

plt.plot([0,v[0]],[0,v[1]],'b',linewidth=2,label='v')
plt.plot([0,Mv[0]],[0,Mv[1]],'r--',linewidth=2,label='Mv')


plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

# Eigenvalues & Eigenvectors

* The algebraic and geometric interpretations of eigenvectors and eigenvalues 
* Hot to perform eigendecomposition in Python (one of the most important concepts in Linear Algebra)
* What "diagonalization" means

# m<span style="color:orange">x</span>m 

Eigendecomposition is defined only for square matrices.

(Singular value decomposition works for any mxn matrix)

# A v = w

# $ \lambda $v = w

### <u>Eigenvalue Equation<u>

# $ \lambda $v = Av

Scaling effect for eigenvectors, where Matrix-vector multiplication behaves similarly to scalar-vector multiplication

### <u>Diagonalization<u>

* A = Matrix
* V = Eigenvectors
* $\Lambda$ = Eigenvalues

# $ \text{AV} = \text{V}\Lambda $

The Eigenvectors are generally accepted as Fullrank and invertible

# $\text{A} =  \text{V} \Lambda \text{V}^{-1}$

```python
M = np.random.randint(-5,6,(5,5))
M = M@M.T

eigvals,eigvecs = np.linalg.eig(M)
fig,ax = plt.subplots(1,3)

ax[0].imshow(M)
ax[0].set_title('The matrix')

ax[1].imshow(np.diag(eigvals))
ax[1].set_title('Eigenvalues')

ax[2].imshow(eigvecs)
ax[2].set_title('Eigenvectors')

plt.show()
```

```python
Mv = M@eigvecs[:,0]
lv = eigvals[0]*eigvecs[:,0]

print(Mv)
print(lv)
```

## <u>Exercise<u>

**Part 1**
1. Extract eigenvectors and eigenvalues from **M** = $ \begin{bmatrix} 1 & 2 \\ 2 & 1 \end{bmatrix} $
2. Plot $\lambda$v and **M**v

```python
M = np.array([ [1,2],
               [2,1] ])

d,V = np.linalg.eig(M)
lv = d[0]*V[:,0]
Mv = M@V[:,0]

plt.plot([0,lv[0]],[0,lv[1]],'b',label='$\lambda v$',linewidth=4)
plt.plot([0,Mv[0]],[0,Mv[1]],'r:',label='Mv',linewidth=4)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.legend()
plt.grid()
plt.show()
```

```python
M = np.array([ [1,2],
               [2,1] ])

d,V = np.linalg.eig(M)
lv = d[1]*V[:,1]
Mv = M@V[:,1]

plt.plot([0,lv[0]],[0,lv[1]],'b',label='$\lambda v$',linewidth=4)
plt.plot([0,Mv[0]],[0,Mv[1]],'r:',label='Mv',linewidth=4)

plt.axis('square')
plt.axis([-3,3,-3,3])
plt.legend()
plt.grid()
plt.show()
```

**Part 2**
1. Create a 6x6 symmetric random-numbers matrix
2. Extract eigenvectors
3. Compute dot products between all pairs of eigenvectors as a matrix

```python
m = 6
Q = np.random.randn(m,m)
Q = Q@Q.T

d,V = np.linalg.eig(Q)
dps = np.zeros((m,m))

for i in range(m):
    for j in range(m):
        dps[i,j] = np.dot(V[:,i],V[:,j])
        

print('\n')        
print(np.round(dps,2))
print('\n')
plt.imshow(dps)
plt.title('Identity Q')
plt.show()
```

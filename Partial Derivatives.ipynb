# $ f(x,y) = 2xy^2 $

# $ \frac{\partial f}{\partial x} = f_x = 2y^2 $

The partial derivative of f, with respect to x

partial f, partial x

# $ \frac{\partial f}{\partial y} = f_y = 4xy^2 $

The partial derivative of f, with respect to y

partial f, partial y

```python
from sympy.abc import x,y
f = x**2 + x*y**2

display(Math('\\frac{\\partial f}{\\partial x} = %s' %(sym.latex(sym.diff(f,x)))))
display(Math('\\frac{\\partial f}{\\partial y} = %s' %(sym.latex(sym.diff(f,y)))))
```

### Exercise

# $ f(x,y) = x^2 + xy^2 $

```python
# implement the function
f = x**2 + x*y**2
dfx = sym.diff(f,x)
dfy = sym.diff(f,y)

# try plotting graphs
plt = sym.plotting.plot3d(f,(x,-3,3),(y,-3,3),title='$f(x,y) = %s$' %(sym.latex(f)))
plt = sym.plotting.plot3d(dfx,(x,-3,3),(y,-3,3),title='$\\frac{\partial f}{\partial x} = %s$' %(sym.latex(dfx)))
plt = sym.plotting.plot3d(dfy,(x,-3,3),(y,-3,3),title='$\\frac{\partial f}{\partial y} = %s$' %(sym.latex(dfy)))
```

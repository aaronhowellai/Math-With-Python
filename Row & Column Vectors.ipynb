# Row & Column Vectors

    The algebraic and geometric definitions of vectors 
    The difference between row and column vectors 
    How to make 3D plots in Python

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math,display
from mpl_toolkits.mplot3d import Axes3D
```

### Algebraic & Geometric Vectors

    Vector: [ an ordered list of numbers ] 
    Dimensionality: The quantity of numbers (elements)

Geometric Vectors can have "tails" and "heads"

Row Vectors represented geometrically do not always correspond to co-ordinates exactly, but when they do, it is referred to the "standard position" 

```python
row_vec = np.array([4,2])
col_vec = np.array([[-2], [3]])
print(row_vec)
print(col_vec)
```

```python
display(Math(sym.latex(sym.sympify(row_vec))))
```

```python
display(Math(sym.latex(sym.sympify(col_vec))))
```

```python
# Create a 2D column vector
col_vec = np.array([[-2], [3]])

plt.grid()
# Plot the column vector using plt.quiver
plt.plot([0,row_vec[0]],[0,row_vec[1]],'r',label='Row Vector')
# plt.plot([0,col_vec[0]],[0,col_vec[1]],'r',label='Row Vector')

plt.quiver(0, 0, col_vec[0], col_vec[1], angles='xy', scale_units='xy', scale=1,color='blue',label='Col Vector')


# Set axis limits and labels

plt.legend()


# Show the plot
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.show()
```

### Exercise

```python
row_vec1 = np.array([3,0,-4])
row_vec2 = np.array([-1,1,3])
print(row_vec1)
print(row_vec2)
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111,projection='3d')


ax.plot([-5,5],[0,0],[0,0],'k--',alpha=.5)
ax.plot([0,0],[-5,5],[0,0],'k--',alpha=.5)
ax.plot([0,0],[0,0],[-5,5],'k--',alpha=.5)

ax.plot([0,row_vec1[0]],[0,row_vec1[1]],[0,row_vec1[2]],'b',linewidth=3,label='row vector 1, %s'%row_vec1)
ax.plot([0,row_vec2[0]],[0,row_vec2[1]],[0,row_vec2[2]],'r',linewidth=3,label='row vector 2, %s'%row_vec2)

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)

plt.title('Plotting of 3-Dimensional Vectors')
plt.legend()
plt.show()
```

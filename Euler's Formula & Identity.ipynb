### $ e^{ix} = cos(x) + isin(x) $

(For points from vector are on the unit circle)
(Where the real axis is x and the imaginary axis is y)

### $ me^{ik} = m(cos(k) + isin(k) ) $

(Where m = magnitude , and k = angle)

### $ e^{i0} = 1 + 0i $

### $ e^{i\pi/2} = 0 + 1i $

### $ e^{i\pi} = -1 + 0i $

### $ e^{i\pi} + 1 = 0 $

```python
k = np.pi /4
m = 2.3

eul = m*np.exp(1j*k)
cis = m*(np.cos(k)+1j*np.sin(k))

print(cis)
print(eul)
print(cis-eul)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

plt.polar([0,ang],[0,mag],'r')
plt.polar(k,m,'bo')
plt.show()
```

```python
def eulerFromCosSin():
    real = eval(input('cosine part: '))
    imag = eval(input('sine part: '))
    
    m = np.sqrt( real**2 + imag**2 )
    k = np.arctan2(imag,real)
    
    plt.polar([0,k],[0,m],'b',linewidth=3)
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$ = %g' %(m,k))
    plt.thetagrids([0,45,130,200,222.2])
    plt.show()
```

### Random Snakes With Cosine & Sine Exercise

```python
theta = np.linspace(0,8*np.pi,1000)
r1 = np.random.rand()
r2 = np.random.rand()
x = np.cos(r1*theta)
y = np.cos(r2*theta)

x = np.append(x,x[0])
y = np.append(y,y[0])

plt.plot(x,y,'k')
plt.title('r$_1$=%s,r$_2$=%s'%(round(r1,2),round(r2,2)))
plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.axis('off')
plt.show()
```

```python
x = np.linspace(-1,1,2000)
y = (1-x**4)**(1/4)

plt.plot(x,y,'k',linewidth=3)
plt.plot(x,-y,'k',linewidth=3)

plt.axis('off')
plt.axis('square')
plt.show()
```

# Logarithmic Spiral

    How to model seashells, hurricanes and galaxies
    How to make a logarithmic spiral
    More art

```python
t = np.linspace(0,10*np.pi,1234)
k = -3
x = np.cos(t)*np.exp(t/k)
y = np.sin(t)*np.exp(t/k)

plt.plot(x,y,'k')
plt.axis('square')
plt.show()
```

# Rose Curves

    Rose Curves and the formula for their petals
    Using powers as parameters in a family of eqautions
    Making curvy diamonds in Python

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
k = 4
t = np.linspace(0,8
    *np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,'m',linewidth=3)
plt.axis('square')
plt.show()
```

# Exercise
Plotting Rose Curves using trigonometric functions in a family of equations

```python
fig,ax = plt.subplots(3,3,figsize=(10,8))
row,col = np.indices((3,3))
t = np.linspace(0,8*np.pi,1000)

for i in range(9):
    k = i/2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('square')
    ax[r,c].axis('off')
    ax[r, c].set_title('k = %g' %k,color='gray')
    
plt.show()    
```

```python
fig,ax = plt.subplots(5,5,figsize=(10,8))
row,col = np.indices((5,5))
t = np.linspace(0,8*np.pi,1000)

for i in range(25):
    k = i/2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'k',linewidth=.5)
    ax[r,c].axis('square')
    ax[r,c].axis('off')
    ax[r, c].set_title('k = %g' %k,color='gray')
    
plt.show()    
```

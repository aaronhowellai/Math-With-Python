```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# Singular Value Decomposition

* The "big picture" of the SVD
* How to perform SVD on a matrix
* How to import pictures into the Python workspace

### SVD: The Bigger Picture

[ A ] 
* The matrix A - **mxn**

[ U ] 
* Orthogonal basis for column space of A - **mxm**

[ $\Sigma$ ] 
* Singular values of A - **mxn**
* Diagonal Matrix

[ $\text{V}^\text{T}$ ] 
* Orthogonal basis for row space of A - **nxn**

```python
ein = Image.open('einstein.jpg')
ein = np.mean(ein,2)

print(np.shape(ein))

plt.imshow(ein)
plt.axis('off')
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)

print(np.shape(U))
print(np.shape(s))
print(np.shape(V))
```

```python
fig,ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title('U')
ax[0].axis('off')

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title('$\Sigma$')
ax[1].axis('off')

ax[2].imshow(V)
ax[2].set_title('$V^{T}$')
ax[2].axis('off')


plt.show()
```

```python
plt.plot(s,'ks-',markerfacecolor='w')
plt.xlim([-1,50])
plt.title('Eigenspectrum of Einstein')
plt.show()
```

```python
randorder = np.random.permutation(len(s))

S = np.zeros(np.shape(ein))
for i in range(0,len(s)):
    S[i,i] = s[i]
    
rein = U@S@V

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD version')

plt.show()
```

```python
# shuffle the order of the singular values 
randorder = np.random.permutation(len(s))

S = np.zeros(np.shape(ein))
for i in range(0,len(s)):
    S[i,i] = s[randorder[i]]
    
rein = U@S@V

plt.subplot(1,2,1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD version')

plt.show()
```

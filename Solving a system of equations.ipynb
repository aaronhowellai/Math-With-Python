# Solving a System of Equations

**In this lecture:**

* The three possibilities for solutions

# $ 2x + y - z = 1 $

# $ 3x + 4y + 2z = 13 $

# $ x-5y-2z=0 $

# $ \begin{bmatrix} 2 & 1 & -1 \\ 3 & 4 & 2 \\ 1 & -5 & -2 \end{bmatrix} \space \space \begin{bmatrix} x \\ y \\ z \end{bmatrix} \space = \space \begin{bmatrix} 1 \\13\\0 \end{bmatrix} $

# $ \text{X}\beta = \text{y} $

### <u>Solve For Beta<u>

**"Least-squares solution"** - works most of the time whether the matrix is square or not 

One of the most important eqautions in all of statistics: regression, ANOVA, and general linear models

# $ \beta = (\text{X}^T\text{X})^{-1}\text{X}^T\text{y} $

```python
X = np.array([ [2,1,-1],[3,4,2],[1,-5,-2] ])
y = np.array([1,13,0])

beta1 = np.linalg.inv(X.T@X)@(X.T@y)
print(beta1)

beta2 = np.linalg.solve(X,y)
print(beta2)
```

```python
# system 1
X1 = np.array([ [3,-1],[-1,1] ])
y1 = np.array([6,2])

# system 2
X2 = np.array([ [3,-1],[1.5,-.5] ])
y2 = np.array([6,3])

# system 3
X3 = np.array([ [3,-1],[1.5,-.5] ])
y3 = np.array([6,2])
```

### System 1

```python
xlim = np.array([0,10])

# convert to y=mx+b
yy1 = -X1[0,0]/X1[0,1]*xlim+y1[0]/X1[0,1]
yy2 = -X1[1,0]/X1[1,1]*xlim+y1[1]/X1[1,1]


plt.plot(xlim,yy1,'b',label='eq 1')
plt.plot(xlim,yy2,'r',label='eq 2')

# solve system
beta = np.linalg.solve(X1,y1)
print(beta)
plt.plot(beta[0],beta[1],'mo',markersize=10,label='$\\beta$')

plt.legend()
plt.grid()
plt.show()
```

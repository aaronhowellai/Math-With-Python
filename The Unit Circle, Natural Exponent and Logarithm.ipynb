# The Unit Circle

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import sympy as sym
import numpy as np
from IPython.display import Math,display
```

```python
x = np.linspace(0,2*np.pi,100)

plt.plot([-1.1,1.1],[0,0],'--',color=[.8,.8,.8])
plt.plot([0,0],[-1.1,1.1],'--',color=[.8,.8,.8])
angle = np.pi/4
plt.plot([0,np.cos(angle)],[0,np.sin(angle)],'r')
plt.plot(np.cos(x),np.sin(x),'k')



plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')
plt.title('The Unit Circle')

plt.show()
```

## Exercise

```python
x = np.linspace(-3,1.83,50)
k = np.exp(x)

for i in k:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis('off')
plt.title('np.exp')
plt.axis('square')
plt.show()
```

# The Natural Exponent & Logarithm

```python
x = np.linspace(-3,3,200)

plt.plot(x,np.exp(x))
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.title('$e^x$')
plt.show()

plt.plot(x,np.log(x))
plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('log(x)')
plt.show()
```

## Exercise

```python
y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

x = np.linspace(.0001,10,100)

for i in y1:
    plt.plot(x,i,'b-')
    for j in y2:
        plt.plot(x,j,'ro')

plt.show()
```

Example

```python
x = np.linspace(.0001,10,10)
y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x,y1,label='$\\log(e^x)$')
plt.plot(x,y2,'o',label='$e^{\\log(x)}$')

plt.axis('square')
plt.legend()
plt.show()
```
